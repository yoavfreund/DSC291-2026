```python
%pylab inline
import numpy as np
from scipy.stats import norm
```

```python
figure(figsize=[10,8])
x = np.linspace(norm.ppf(0.001),
                norm.ppf(0.999), 100)

rv = norm()
plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

rv = norm(loc=0.3)
plot(x, rv.pdf(x), 'r-', lw=2, label='frozen pdf')
title('repeat experiment once')
grid()
```

```python
figure(figsize=[10,8])
x = np.linspace(norm.ppf(0.001),
                norm.ppf(0.999), 1000)

rv = norm(scale=0.1)
plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

rv = norm(loc=0.3,scale=0.1)
plot(x, rv.pdf(x), 'r-', lw=2, label='frozen pdf')
title('repeat experiment 100 times')
grid()
```

```python
figure(figsize=[10,8])
x = np.linspace(norm.ppf(0.001),
                norm.ppf(0.999), 1000)

rv = norm(scale=0.01)
plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')

rv = norm(loc=0.3,scale=0.01)
plot(x, rv.pdf(x), 'r-', lw=2, label='frozen pdf')
title('repeat experiment 10,000 times')
grid()
```
